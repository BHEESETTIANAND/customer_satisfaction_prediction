```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
x=df.drop("satisfaction",axis=1)
y=df["satisfaction"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import mlflow
```

```python
mlflow.set_experiment("satisfaction_prediction")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint

pipelines = {
    "DT": Pipeline([
        ("classifier", DecisionTreeClassifier())
    ]),
    "RF": Pipeline([
        ("classifier", RandomForestClassifier())
    ])
}

param_grids = {
    "DT": [{
        'classifier__criterion': ['gini', 'entropy'],
        'classifier__max_depth': [5, 10, 15, 20]
    }],
    "RF": [{
        'classifier__criterion': ['gini', 'entropy'],
        'classifier__max_depth': [5,10,15,10]
    }]
}

```

```python
import time
import os
import joblib
best_models={}
for i in pipelines.keys():
    print("*"*10,i,"*"*10)
    grid_search=GridSearchCV(estimator=pipelines[i],
                             param_grid=param_grids[i],
                             cv=5,
                             scoring="accuracy",
                             return_train_score=True,
                             verbose=1)
    
    mlflow.sklearn.autolog(max_tuning_runs=None)
    with mlflow.start_run() as run:
         grid_search.fit(x_train,y_train)
    print("train score",grid_search.best_score_)
    print("test score",grid_search.score(x_test,y_test))
    best_models[i]=grid_search.best_estimator_
    print()
    model_dir="best_models"
    os.makedirs(model_dir,exist_ok=True)
    model_path=os.path.join(model_dir,f"{i}.pkl")
    joblib.dump(grid_search.best_estimator_,model_path)
    model_size=os.path.getsize(model_path)
```
